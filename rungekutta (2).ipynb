```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.02
a=1000.

tlist = []
wlist = []
vlist = []

d= {'t': tlist,'w': wlist , 'v': vlist}
i=0
t = 0
w = 1
v = 0
tlist.append(t)
wlist.append(w)
vlist.append(v)
def rungekutta():
    global t
    global i
    global w
    global v
    k1 = h*(-w)*v
    l1 = h*w*w
    k2 = h*(-(w+(k1/2)))* (v+(l1/2))
    l2 = h*((w+(k1/2)))**2
    k3 = h*(-(w+(k2/2)))* (v+(l2/2))
    l3 = h*((w+(k2/2)))**2
    k4 = h*(-(w+k3))* (v+l3)
    l4 = h*(w+k3)**2
    w = w + (k1+2*k2+2*k3+k4)/6
    v = v + (l1+2*l2+2*l3+l4)/6
    t+=h
    tlist.append(t)
    wlist.append(w)
    vlist.append(v)

while i<a:
    rungekutta()
    i+=1
print(d)
```
